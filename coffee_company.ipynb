## 1. Importing the Libraries

Python libraries like NumPy, Pandas, Matplotlib, and Seaborn are needed for data cleaning, data analysis, and visualization.

```python
#Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

## 2. Read and summarize the dataset

```python
#Read the CSV file stored in the same folder as the notebook
cc = pd.read_csv("Coffee_company.csv")

cc.head(20)
```

```python
# Summarizing the info 

cc.info()
```

We can identify that there are no null values in all 1575 entries

## 3. Data Cleaning

To identify Null values

```python
cc.isnull()
```

```python
# Finding no of null values in each column

cc.isnull().sum()
```

No null values found

```python
# Dropping unwanted columns
cc = cc.drop(columns = [" Discount Band "])

cc. head()

```

```python
cc.columns
```

Here We can see that the column headings have spacing in the front and back make it hard to call them. 

```python
#Deleting the spaces in the columns

cc.columns = cc.columns.str.strip()
cc.columns
```

Here again we can see that most of the data is in object datatype whereas we need them in float64 or int64 for calculations.

```python
#Converting specific columns to numeric values

# List of columns with $ sign
columns_to_convert = ['Manufacturing Price', 'Sale Price', 'Gross Sales', 'Discounts',
       'Sales', 'COGS', 'Profit']

#Converted the values into strings and removed all the unwanted characters, then converted to numeric
for i in columns_to_convert:
    cc[i] = cc[i].astype(str).str.replace('$', '').str.strip()
    cc[i] = cc[i].astype(str).str.replace(',', '').str.strip()
    cc[i] = cc[i].astype(str).str.replace('-', '').str.strip()
    cc[i] = cc[i].astype(str).str.replace('(', '').str.strip()
    cc[i] = cc[i].astype(str).str.replace(')', '').str.strip()
    cc[i] = cc[i].astype(str).str.replace('nan', '0').str.strip()
    cc[i] = pd.to_numeric(cc[i])

cc.dtypes
print(cc)
```

```python
cc.info()
```

Upon Closer look we can identify that the units sold are in decimals, Which is physically not possible.

```python
#Rounding up the Units Sold Values
cc["Units Sold"] = round(cc["Units Sold"]).astype(int)
cc["Units Sold"]
```

```python
print(cc["Date"].dtype)
```

We can see that the Date is in "object" datatype. Lets change that into Date format

```python
cc["Date"] = pd.to_datetime(cc["Date"])

print(cc["Date"])
```

```python
# Dropping more unwanted columns
cc = cc.drop(columns = ["Month Number","Month Name","Year"])

cc.head()
```

## 4. Data Manipulation

Lets now group together based on segments and countries to get more insights

```python
#Grouping the dataset based on their segments

cc_segments = cc[["Segment", "Units Sold", "Sales", "Profit"]]
cc_segments = cc_segments.groupby(by = ["Segment"]).sum().reset_index()

cc_segments.head()
```

```python
#Grouping the dataset based on Country of origin

cc_country = cc[["Country", "Units Sold", "Sales", "Profit"]]
cc_country = cc_country.groupby(by = ["Country"]).sum().reset_index()

cc_country.head()
```

```python
#Grouping the dataset based on Product

cc_product = cc[["Product", "Units Sold", "Sales", "Profit"]]
cc_product = cc_product.groupby(by = ["Product"]).sum().reset_index()

cc_product.head()
```

## 5. Data Visualization

We use Matplotlib to plot Bar Charts to show the visualization

```python
#Creating a grouped Bar Graph for the different segments

segment = cc_segments["Segment"].tolist()
y1 = cc_segments["Units Sold"].tolist()
y2 = cc_segments["Sales"].tolist()
y3 = cc_segments["Profit"].tolist()

font1 = {
    "family" : "sans serif",
    "color"  : "black",
    "size"   : 15
}
font2 = {
    "family" : "sans serif",
    "color"  : "darkred",
    "size"   : 10
}

x = np.arange(len(segment))
width = 0.2

fig, ax = plt.subplots()

p1 = ax.bar(x, y1, width)
p2 = ax.bar(x+width, y2, width)
p3 = ax.bar(x+(2*width), y3, width)

plt.xticks(x, segment, rotation = 90)

plt.title("Grouped chart on Segments", fontdict = font1)
plt.xlabel("Segment", fontdict = font2)
plt.ylabel("Units Sold, Sales, Profit", fontdict = font2)
plt.show()
```
